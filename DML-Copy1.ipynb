## Insert data into the tables we have created from both csv files(imported and url)

```python
%%sql 
INSERT INTO Team(TeamID, TeamName)
SELECT DISTINCT TeamID, TeamName
From NBAData;

INSERT INTO Game(GameNumber, TeamId, Minutes, Year)
SELECT DISTINCT GameID, Team.TeamID, Minutes, Year
FROM Team JOIN NBAData ON (NBAData.TeamID = Team.TeamID);


INSERT INTO Stats(Points, TeamID, ThreePts, FreeThrow, OffenseRebound, DefenseRebound, Home, Win)
SELECT DISTINCT Points, Team.TeamID, ThreePts, FreeThrow, OffensiveRebound, DefensiveRebound, Home, Win
From Team JOIN NBAData ON (NBAData.TeamID = Team.TeamID);

INSERT INTO Season(Year, TeamID)
SELECT DISTINCT Year, Team.TeamID
FROM NBAData Join Team ON (Team.TeamID = NBAData.TeamID);

INSERT INTO Player(PlayerName, TeamID, SeasonID, Age)
SELECT DISTINCT Player, Team.TeamID, Season.SeasonID, Age
FROM PlayerData JOIN Team ON (Team.TeamID = PlayerData.Tm)
            JOIN Season ON (Season.Year = PlayerData.year);
```

```python
%%sql
SELECT DISTINCT Player, Season.TeamID, Year
from Player JOIN Season ON (Player.TeamID=Season.TeamID)
where Season.Year = 2000
;
```

**Let's find out the average age of players on the winning team in 1990**

```python
%%sql 
SELECT AVG(AGE),TeamName, Season.Year
from Player JOIN Team ON (Player.TeamID=Team.TeamID)
            JOIN SEASON ON (Player.SeasonID=Season.SeasonID);
```

**Let's find out how many times the Celtics has won a championship since 1980**

```python
%%sql
SELECT COUNT(DISTINCT YEAR), TeamName
From Season JOIN Team ON (Team.TeamID=Season.TeamID)
Where Team.TeamID = 'BOS';
```

**....and who has won the championships the most since 1980..**

```python
%%sql
SELECT Count(Year), TeamName
From Season JOIN Team ON (Team.TeamID=Season.TeamID)
Group by TeamName
Order by Count(year) desc;
```
